```python
# 파일 생성
%%writefile app.py
import streamlit as st
import pandas as pd

st.title('고객 이탈 예측')

st.subheader('데이터셋을 업로드하세요')
upload_file = st.file_uploader("CSV 파일을 업로드", type=["CSV"])

if upload_file is not None:
    data = pd.read_csv(upload_file)
    st.write(data.head())
```
